```python
train_data = data.copy()
train_data['diagnosis'] = train_data['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
train_data = train_data.drop(['id'], axis=1)
```

```python
X=data.drop('diagnosis',axis=1).values
y=data['diagnosis'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1234)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
accuracy_list=[]
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000,verbose=0)
lr.fit(scaled_X_train, y_train)
y_pred = lr.predict(scaled_X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
#accuracy_list.append(['LogRegression',accuracy])
```

```python
accuracy_list
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200, random_state=1234)
model.fit(scaled_X_train, y_train)
predictions = model.predict(scaled_X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
accuracy_list.append(['RandomForest',accuracy])
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(scaled_X_train,y_train)
predictions = svc.predict(scaled_X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
accuracy_list.append(['SupportVectorClassifier',accuracy])
```

```python
print("Accuracy")
for i in accuracy_list:
  print(i[0],"-->",i[1])
```
